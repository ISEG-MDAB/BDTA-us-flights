```
from geopy.geocoders import Nominatim
```

```
df = spark.table("complete_dataset")
display(df)
```

```
airport_info = spark.table("airport_id").toPandas()
airport_info.head()
```

```
used_airports = df.select("ORIGIN_AIRPORT_ID").distinct().union(df.select("DEST_AIRPORT_ID").distinct()).distinct().toPandas()
airport_info['Code'] = airport_info['Code'].astype('int32')
used_airports.head()
```

```
import pandas as pd
airports = pd.merge(used_airports, airport_info, left_on="ORIGIN_AIRPORT_ID", right_on="Code", how='inner')
airports = airports.drop('ORIGIN_AIRPORT_ID', axis = 1)
airports[['AIPORT_LOC','AIRPORT_NAME']] = airports['Description'].str.split(':', expand = True)
airports = airports.drop('Description', axis = 1)
airports =  airports.sort_values('Code')
airports = airports.reset_index(drop=True)
airports['LAT'] = ''
airports['LON'] = ''
airports.head()
```

```
#LOC cleaning
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Sun Valley/Hailey/Ketchum, ID', 'Sun Valley, Hailey, ID')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Harlingen/San Benito, TX', 'San Benito, TX')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Saginaw/Bay City/Midland, MI', 'Saginaw,Bay City,Midland, MI')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Bristol/Johnson City/Kingsport, TN', 'Bristol, Johnson City, TN')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Mission/McAllen/Edinburg, TX', 'McAllen, Edinburg, TX')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Newburgh/Poughkeepsie, NY', 'Newburgh, NY')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Riverton/Lander, WY', 'Riverton, WY')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('New Bern/Morehead/Beaufort, NC', 'New Bern, Beaufort, NC')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Newport News/Williamsburg, VA', 'Newport News, VA')
airports['AIPORT_LOC'] = airports['AIPORT_LOC'].replace('Iron Mountain/Kingsfd, MI', 'Iron Mountain, MI')



#Use geolacator to get the coordinates from the airport location
geolocator = Nominatim(user_agent="andreviana@aln.iseg.ulisboa.pt")
for i in range(len(airports)):
  location = geolocator.geocode(airports['AIPORT_LOC'][i])
  airports['LAT'][i] = location.latitude
  airports['LON'][i] = location.longitude
```

```
#Rename this airport because there are 2 airports with the same name
airports['AIRPORT_NAME'][airports.index[airports['Code'] == 15323].tolist()[0]] = 'Tri Cities Kingsport'
```

```
from pyspark.sql import SparkSession
airports = spark.createDataFrame(airports)
```
