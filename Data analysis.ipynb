```
from datetime import datetime
import pandas as pd
import plotly.graph_objects as go
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1) Load data

```
flights_data = spark.table("flights_data_agg").toPandas()
flights_data['FL_DATE'] = pd.to_datetime(flights_data['FL_DATE'])
flights_data.head()
```

```
network_data = spark.table("network_data").toPandas()
network_data.head()
```

```
airport_info = spark.table("airports_info").toPandas()
airport_info.set_index('Code', inplace = True)
airport_info.head()
```

#2) Plot with the number of flights per airport

```
#Create a dataframe where the data is grouped by airport and month
total_airport = flights_data[['FL_DATE','ORIGIN_AIRPORT_ID','TOTAL_FLIGHTS']].copy()
total_airport["id"] = total_airport['ORIGIN_AIRPORT_ID'].astype(str) + ':' + total_airport["FL_DATE"].dt.strftime('%Y-%m').astype(str)
total_airport = total_airport.drop('ORIGIN_AIRPORT_ID', axis = 1)
total_airport = total_airport.groupby('id').sum()
total_airport.reset_index(inplace = True)
total_airport[['value','level']] = total_airport['id'].str.split(':', 1, expand=True)

#Create a dataframe with dates 2017-01 to 2020-12 in rows and airportsID in column
aer = pd.DataFrame(columns=flights_data['ORIGIN_AIRPORT_ID'].unique())
datelist = pd.DataFrame(pd.date_range(start='1/2017', end='01/2021', freq = 'M'), columns = ['DATE'])
datelist['DATE'] = datelist['DATE'].dt.strftime('%Y-%m')
aer['DATES'] = datelist['DATE']
aer.set_index('DATES', inplace=True)

#Fill the aer dataframe with the total_aiport data
for value in aer.columns:
  for level in aer.index:
    v = str(value)
    aux = total_airport[(total_airport['level'] == level) & (total_airport['value'] == v)]['TOTAL_FLIGHTS'].values
    for i in aux: aer[value][level] = i

aer = aer.fillna(0)

#Rename and order the columns by the Name
airport_name = airport_info['AIRPORT_NAME'].to_dict()
aer = aer.rename(columns = airport_name)
aer = aer.reindex(sorted(aer.columns), axis=1)
#Calculate the avg of fligts per airport
aer_avg = aer.mean(axis=1)

aer.head()
```

```
fig = go.Figure()

fig.add_trace(go.Scatter(x=aer.index, y=aer['Aberdeen Regional'],  line=dict(color='rgb(40, 150, 150)'), name='Airport Aberdeen Regional'))

fig.add_trace(go.Scatter(x=aer.index, y=aer_avg, name='Avarage per Airport'))

fig.update_layout(title=dict(
  text='Number of flights per month, per airport',
  x=0.5,
  font = dict(
    family='Arial',
    color='black')),
  xaxis=dict(
    title=dict(
      text='Year',
      font=dict(
        family='Arial',
        size=14,
        color='black')),
    showline=True,
        showgrid=True,
        showticklabels=True,
        ticks='outside',
        linewidth=2,
        linecolor='rgb(204,204,204)'),
  yaxis=dict(
    title=dict(
      text='Number of flights',
      font=dict(
        family='Arial',
        color='black')),
    ticks='outside',
        linewidth=2,
        linecolor='rgb(204,204,204)'),
    font=dict(
        family='Arial',
        color='black'))

buttons = []
# add buttons for each type of data

for airport in aer.columns:
    buttons.append(dict(method='restyle',
                        label=airport,
                        visible=True,
                        args=[{'y':[aer[airport]],
                            'x':[aer.index],
                              'type':'scatter',
                               'name':'Airport '+ str(airport)
                              }, [0]]))
    
# Adjust dropdown placement
button_layer_1_height = 1.2
updatemenus = list([
    dict(buttons=buttons,
        direction = 'down',
        pad = {'r':10, 't':17},
        showactive = True,
        x = 0,
        xanchor = 'left',
        y = button_layer_1_height,
        yanchor='top')])


fig.update_layout(updatemenus=updatemenus)
```

#3) Plot DataFrame with the highest and lowest numbers of flights per month

```
# Create a data frame with the airports with 3 highest number of flights per month
c1 = ['Highest n. of flights', '2nd highest n. of flights', '3rd highest n. of flights']
df_max = (aer.apply(lambda x: pd.Series(x.nlargest(3).index, index=c1), axis=1))

# Create a data frame with the airports with 3 lowest number of flights per month
c2 = ['Lowest n. of flights', '2nd lowest n. of flights', '3rd lowest n. of flights']
df_min = (aer.apply(lambda x: pd.Series(x.nsmallest(3).index, index=c2), axis=1))

# Merge the dataframes
df_months = df_max.merge(df_min, left_index=True, right_index=True)
```

#4) Plot the number of canceled flights, delayed flights and average delay per state

```
#Create a dataframe where the data is grouped by state and year
total_state = flights_data[['FL_DATE','ORIGIN_STATE_ABR','TOTAL_FLIGHTS','CANCELLED','TOTAL_DEP_DELAY','DEP_DELAY_LESS15','DEP_DELAY_MORE15']].copy()
total_state["id"] = total_state ['ORIGIN_STATE_ABR'] + ':' + total_state ["FL_DATE"].dt.strftime('%Y').astype(str)
total_state = total_state.drop('ORIGIN_STATE_ABR', axis = 1)
total_state = total_state.groupby('id').sum()
total_state.reset_index(inplace = True)

avg_state = flights_data[['FL_DATE','ORIGIN_STATE_ABR','AVG_DELAY']].copy()
avg_state["id"] = avg_state ['ORIGIN_STATE_ABR'] + ':' + avg_state ["FL_DATE"].dt.strftime('%Y').astype(str)
avg_state = avg_state.drop('ORIGIN_STATE_ABR', axis = 1)
avg_state = avg_state.groupby('id').mean()
avg_state.reset_index(inplace = True)

total_state = pd.merge(total_state, avg_state[['AVG_DELAY','id']], on="id", how='inner')
total_state[['state','year']] = total_state['id'].str.split(':', 1, expand=True)

#Create a empty dataframe
tot = pd.DataFrame(columns=['TOTAL_FLIGHTS', 'CANCELLED', 'TOTAL_DEP_DELAY', 'DEP_DELAY_LESS15', 'DEP_DELAY_MORE15','AVG_DELAY'])
tot['State'] = flights_data['ORIGIN_STATE_ABR'].value_counts().index
tot.set_index('State', inplace=True)

# Create a dictionary with the years as keys and as values a data frame with columns 'CANCELLED', 'TOTAL_DEP_DELAY', 'DEP_DELAY_LESS15', 'DEP_DELAY_MORE15', 'AVG_DELAY'
tot_years = dict()
years = [2017,2018,2019,2020]

for year in years:
  tot_years[year] = tot.copy()
  for state in tot_years[year].index:
    for value in tot_years[year].columns:
      aux = total_state[(total_state['year'] == str(year)) & (total_state['state'] == state)][value]
      for i in aux: tot_years[year][value][state] = i
```

```
fig = go.Figure(data=go.Choropleth(
    locations = tot.index,
    z = tot_years[2017]['CANCELLED'].astype(float), # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = 'Reds',
    colorbar = dict(
        title = dict(
            text="Count",
            font=dict(
                family='Arial',
                size=14,
                color='black')),
        tickfont=dict(
            family='Arial',
            color='black',
            size=14)),
    marker_line_color='white',
))

fig.update_layout(
    title = dict(
        text = 'Number of Flights / Cancellations / Delays / Average Delay per year',
        x=0.5,
        font=dict(
            family='Arial',
            color='black',
            size=20)),
    geo_scope='usa', # limite map scope to USA
    legend=dict(
        font=dict(
            family='Arial',
            color='black',
            size=16),
        bgcolor='lavender',
        borderwidth=1)

)

names = dict(TOTAL_FLIGHTS = 'Total Flights',
            CANCELLED = 'Cancellations',
            TOTAL_DEP_DELAY = 'Flights with delay',
            DEP_DELAY_LESS15 = 'Flights with delay < 15min',
            DEP_DELAY_MORE15 = 'Flights with delay >= 15min',
            AVG_DELAY = 'Average Delay')

# Add buttons
buttons_1 = []


for year in tot_years.keys():
  for level in tot_years[year].columns:
    buttons_1.append(dict(method = 'update',
                         args = [{'z':[tot_years[year][level]]}],
                         label = str(year) + ' ' + names[level]))

    
# Adjust dropdown placement
updatemenus = list([
    dict(buttons=buttons_1,
        direction = 'down',
        pad = {'r':1, 't':10},
        showactive = True,
        x = 0)])
   
fig.update_layout(updatemenus=updatemenus)
```

#5) Plot the rate of canceled flights and delayed flights per state

```
#Create a dataframe where the data is grouped by state and year
rate_state = flights_data[['FL_DATE','ORIGIN_STATE_ABR','CANCELLED','TOTAL_DEP_DELAY','DEP_DELAY_LESS15','DEP_DELAY_MORE15']].copy()
rate_state["id"] = rate_state ['ORIGIN_STATE_ABR'] + ':' + rate_state ["FL_DATE"].dt.strftime('%Y').astype(str)
rate_state = rate_state.drop('ORIGIN_STATE_ABR', axis = 1)
rate_state = rate_state.groupby('id').sum()
rate_state.reset_index(inplace = True)

rate_state[['state','year']] = total_state['id'].str.split(':', 1, expand=True)

#Create a empty dataframe
rate = pd.DataFrame(columns=['CANCELLED', 'TOTAL_DEP_DELAY', 'DEP_DELAY_LESS15', 'DEP_DELAY_MORE15'])
rate['State'] = flights_data['ORIGIN_STATE_ABR'].value_counts().index
rate.set_index('State', inplace=True)

# Create a dictionary with the years as keys and as values a data frame with columns 'CANCELLED', 'TOTAL_DEP_DELAY', 'DEP_DELAY_LESS15', 'DEP_DELAY_MORE15', 'AVG_DELAY'
rate_years = dict()
years = [2017,2018,2019,2020]

for year in years:
  rate_years[year] = rate.copy()
  for state in rate_years[year].index:
    for value in rate_years[year].columns:
      aux = rate_state[(rate_state['year'] == str(year)) & (rate_state['state'] == state)][value]
      for i in aux: rate_years[year][value][state] = i / tot_years[year]['TOTAL_FLIGHTS'][state]
```

```
fig = go.Figure(data=go.Choropleth(
    locations = tot.index,
    z = rate_years[2017]['CANCELLED'].astype(float), # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = 'Blues',
    colorbar = dict(
        title = dict(
            text="Rate",
            font=dict(
                family='Arial',
                size=14,
                color='black')),
        tickfont=dict(
            family='Arial',
            color='black',
            size=14)),
    marker_line_color='white',
))

fig.update_layout(
    title = dict(
        text = 'Rate of Cancellations / Delays',
        x=0.5,
        font=dict(
            family='Arial',
            color='black',
            size=20)),
    geo_scope='usa', # limite map scope to USA
    legend=dict(
        font=dict(
            family='Arial',
            color='black',
            size=16),
        bgcolor='lavender',
        borderwidth=1)

)

names = dict(CANCELLED = 'Cancellations',
            TOTAL_DEP_DELAY = 'Flights with delay',
            DEP_DELAY_LESS15 = 'Flights with delay < 15min',
            DEP_DELAY_MORE15 = 'Flights with delay >= 15min')

# Add buttons
buttons_1 = []


for year in rate_years.keys():
  for level in rate_years[year].columns:
    buttons_1.append(dict(method = 'update',
                         args = [{'z':[rate_years[year][level]]}],
                         label = str(year) + ' ' + names[level]))

    
# Adjust dropdown placement
updatemenus = list([
    dict(buttons=buttons_1,
        direction = 'down',
        pad = {'r':1, 't':10},
        showactive = True,
        x = 0)])
   
fig.update_layout(updatemenus=updatemenus)
```

#6) Network from the routes

```
G = nx.Graph()
for origin in set(network_data['ORIGIN_AIRPORT_ID']):
  G.add_node(origin)

for i in range(len(network_data['ORIGIN_AIRPORT_ID'])):
   G.add_edge(network_data['ORIGIN_AIRPORT_ID'][i], network_data['DEST_AIRPORT_ID'][i], length = network_data['DISTANCE'][i])
```

```
nx.draw_networkx(G)
plt.show()
```

```
# Get node position dict
x, y = airport_info['LON'].values, airport_info['LAT'].values

pos_dict = {}
for index, apt in enumerate(airport_info.index):
    pos_dict[apt] = (x[index], y[index])
for apt, coordinate in pos_dict.items():
    G.nodes[apt]['pos'] = coordinate #nx graph nodes has a property called 'pos'     
```

```
#Add coordinates to the network
nx.draw_networkx_nodes(G=G, pos=pos_dict, node_color='royalblue',
                       alpha=0.8)
nx.draw_networkx_edges(G=G, pos=pos_dict, edge_color='red', alpha=0.2, arrows=False)
```
